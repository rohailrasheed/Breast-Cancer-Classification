# Deleting Columns

```python
df.info() # It will provide all the information related to our data set (.csv file)
```

```python
# plotting a graph using seaborn.
plt.figure(figsize=(10,10))
sns.heatmap(df.isnull()) # Note consider the white portion as a null.
```

```python
# To find the percantage of our data is present or null.

# df.isnull().sum() ----> It will provide sum of all the columns individually.
# df.isnull().sum()/ df.shape[0]  -----> as we want only rows so we take 0 for rows.
# df.isnull().sum()/ df.shape[0] * 100 -----> is used for finding percenatge.


null_var = df.isnull().sum()/ df.shape[0] * 100
null_var
```

```python
# Now we are extracting only those missing values which are greater than 80% missing values.

# null_var[null_var > 80].keys()

# Now we are finding that columns those have missing values greater than 80%

drop_col = null_var[null_var > 80].keys()
drop_col
```

```python
# Now we are dropping these columns from our data set
data_frame_2clm = df.drop(columns=drop_col)
```

# Split the data into X and Y
## We have to convert the our data into matrix & vector because we have independent variable we will converted them into matrix and dependent variable converted into vector.

```python
# Converting the our data into matrix and vector.

# We will drop the diagnosis variable because it is the dependent variable
X = df6_drop_colmns.drop("diagnosis", axis = 1) # For matrix

# Because It is our dependent variable
y = df6_drop_colmns["diagnosis"] # For Vector

print("The shape of X :: ",X.shape)
print("The shape of y :: ",y.shape)
```

# Split the dataset into train and test dataset
Now,Split the dataset into train and test dataset.Because on train dataset we train the ML Algo and on test dataset,we test the
ML Algo.So for this purpose we import train_test_split class from sklearn library.

```python
# Importing Libraries using Sklearn for train and split of data set
from sklearn.model_selection import train_test_split
```

```python
# Now we are splitting our data into train and test.

# test_size = 0.2 means 20% data is for testing and the rest of 80% is used for training.

# train_test_split class will give us 4 data frames i.e. X_train,X_test,y_train,y_test.

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3,random_state = 0)

print("The shape of X_train :: ",X_train.shape)
print("The shape of X_test :: ",X_test.shape)
print("The shape of y_train :: ",y_train.shape)
print("The shape of y_test :: ",y_test.shape)

#Now our data has been split into X_train,X_test,y_train,y_test.
```

# Note:
#### In the output,we can see that,we have different different indexes present,because train_test_split function splits the data randomly for train and test data.It returns the shuffle records.Whenever we run it,it gives different different data. But if you want to keep the data constant for training and testing dataset.Use random_state parameter in train_test_split.

# Phase#4 Feature Scalling
### There is no any thumb rule to use Standardization or Normalization for specific Machine learning algorithm.We will mostly use Standardization (standard scalling) for clustering analyses, principal component analysis (PCA) so that we can get best accuracy result. While Normalization (Min Max Scaler) is mostly  prefer for image processing, neural network algorithm require data in scale 0-1, K-Neighbor Nearest.

# Here We will perform practical implementation mainly on Standard Scaler

```python
# Importing Library using Sklearn for standardization
from sklearn.preprocessing import StandardScaler
```

```python
# Now data is ready for feature scaling. Now using standardscaler for standardization.
standardsc = StandardScaler()
```

```python
# Now Fitting train dataset i.e. X_train.

# standardscaler will learn the mean value and standard deviation from X_train dataset.
standardsc.fit(X_train)
```

```python
standardsc.mean_ # this is the mean value of our features.
```

```python
standardsc.scale_ # this is the standard deviation of our features.
```

```python
X_train.describe()
```

```python
"""
Now the standard deviation object's value which it learns,
we will tranform it on the X_train and X_test dataset.
"""
X_train_standardsc = standardsc.transform(X_train)
X_test_standardsc = standardsc.transform(X_test)
```

### Note: There is no need to scale the dependent variable i.e. diagnosis.

```python
X_train_standardsc
```

```python
X_test_standardsc
```

```python
# Now we are converting our data into data frame.

X_train_standardsc = pd.DataFrame(X_train_standardsc, columns = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean'])
X_test_standardsc = pd.DataFrame(X_test_standardsc, columns = ['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean'])
```
